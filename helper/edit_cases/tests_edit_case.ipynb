```python
def get_mandatory_refs_identifier(obj_type):
    obj = STUDY.create(obj_type)
    obj_dict = obj.as_dict()
    refs = []
    for property in obj_dict.keys():
        if obj.description(property).is_reference():
            refs.append[property]
    
    return refs

def register_name(obj,obj_type,name,comments):
    # Name property
    if name and obj.has_name:
        obj.name = name 
        comments += f"Name: {obj.name}\n"
    elif not name and obj.has_name:
        obj.name = f"{obj_type}_{uuid.uuid4().hex[:6]}"
        comments += f"Automatic Name: {obj.name}\n "
    return comments

def register_code(obj,obj_type,code,comments):
    if code and obj.has_code:
        obj.code = code
        comments += f"Code: {obj.code}\n"
    elif not code and obj.has_code:
        last_code = STUDY.find(obj_type)[-1].code
        obj.code = last_code +1
        comments += f"Automatic Code: {obj.code}\n"
    return comments

def register_id(obj,id,comments):
    if id and obj.has_id:
        obj.id = id 
        comments += f"ID: {obj.id}\n"
    elif not id and obj.has_id:
        obj.id = f"{uuid.uuid4().hex[:2]}"
        comments += f"Automatic id: {obj.id}\n "
    return comments  

def register_keys(obj, key, comments):
    if key:
        obj.key = key
        comments += f"Key: {obj.key}\n"
    else:
        obj.key = f"{obj.name} [{obj.code}]\n"
        comments += f"Automatic Key: {obj.key}\n"
    return comments


def create_element(obj_type:str, name:str= None, code:int =None, key:str=None, id=None,  properties:dict={}):

    obj = STUDY.create(obj_type)
    comments = f"Object of type {obj_type} created\n"

    # Basic properties
    comments = register_name(obj, obj_type,name,comments)
    comments = register_code(obj,obj_type,code,comments)
    comments = register_keys(obj, key, comments)
    comments = register_id(obj,id,comments)

    # Set properties
    for property_name in obj.descriptions().keys():
        property_description = obj.description(property_name)
        if property_name in properties.keys():

            # Set defined references 
            if property_description.is_reference():
                ref_obj_key = properties[property_name]

                # List
                if isinstance(ref_obj_key,list):
                    # case where the user gave a list of objects 
                    obj.set(property_name,ref_obj_key)
                else: 
                    if "List" in property_description.type_description():
                        obj.set(property_name,[ref_obj_key])
                        comments += f"Reference {property_name}: [{ref_obj_key}]\n"
                    else:
                        obj.set(property_name,ref_obj_key)
                        comments += f"Reference {property_name}: {ref_obj_key}\n"

            # Set other properties
            else: 
                try: 
                    value = properties[property_name]
                    obj.set(property_name,value)
                    comments += f"Property {property_name}: {value}"
                except:
                    continue

        
        # Set default references
        elif property_description.is_reference() and property_description.is_required():
            try: 
                desc_type = property_description.type_description()
                if "List" in desc_type:
                    ref_obj_type = desc_type.split()[-1][:-1] # "List of DataObject of type(s): Fuel" -> Fuel
                    ref_obj = STUDY.find(ref_obj_type)[0]
                    obj.set(property_name,[ref_obj])
                    comments += f"Default Reference {property_name}: [{ref_obj}]\n"
                else:
                    ref_obj_type = desc_type.strip() # System 
                    ref_obj = STUDY.find(ref_obj_type)[0]
                    obj.set(property_name,ref_obj)
                    comments += f"Default Reference {property_name}: {ref_obj}\n"
            except Exception as e: 
                print(f"{e}")
                continue


        
    print(comments)        
    STUDY.add(obj)
    return comments
```

```python
import uuid
import logging

# Configure logging to provide feedback
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')
logger = logging.getLogger(__name__)

def get_mandatory_refs_identifier(obj_type: str) -> list:
    """Returns a list of properties that are mandatory references."""
    try:
        obj = STUDY.create(obj_type)
        return [
            prop for prop in obj.as_dict().keys() 
            if obj.description(prop).is_reference()
        ]
    except Exception as e:
        logger.error(f"Failed to identify mandatory refs for {obj_type}: {e}")
        return []

def _handle_basic_properties(obj, obj_type: str, name: str, code: int, key: str):
    """Internal helper to set Name, Code, and Key with fallback logic."""
    
    # Handle Name
    if obj.has_name:
        obj.name = name if name else f"{obj_type}_{uuid.uuid4().hex[:6]}"
        logger.info(f"Assigned Name: {obj.name}")

    # Handle Code
    if obj.has_code:
        if code is not None:
            obj.code = code
        else:
            try:
                existing_elements = STUDY.find(obj_type)
                obj.code = (existing_elements[-1].code + 1) if existing_elements else 1
            except Exception:
                obj.code = 1
        logger.info(f"Assigned Code: {obj.code}")

    # Handle Key
    if key:
        obj.key = key
        logger.info(f"Assigned Key: {obj.key}")

def _set_reference_property(obj, prop_name: str, prop_desc, value):
    """Internal helper to handle logic for reference types (Lists vs Single)."""
    is_list_type = "List" in prop_desc.type_description()
    
    if is_list_type:
        formatted_value = value if isinstance(value, list) else [value] # Format to list if necessary 
    else:
        formatted_value = value[0] if isinstance(value, list) else value # Format to value if necessary
    obj.set(prop_name, formatted_value)
    logger.info(f"Set Reference [{prop_name}]: {formatted_value}")

def create_element(obj_type: str, name: str = None, code: int = None, key: str = None, properties: dict = None):
    """
    Primary tool for the LLM to create objects in the STUDY environment.
    """
    properties = properties or {}
    
    try:
        # 1. Initialization
        obj = STUDY.create(obj_type)
        logger.info(f"Successfully initialized object of type: {obj_type}")

        # 2. Set Basic Identity Properties
        _handle_basic_properties(obj, obj_type, name, code, key)

        # 3. Process All Properties
        all_descriptions = obj.descriptions()
        
        for prop_name, prop_desc in all_descriptions.items():
            # Property provided by the User
            if prop_name in properties:
                value = properties[prop_name]
                if prop_desc.is_reference():
                    _set_reference_property(obj, prop_name, prop_desc, value)
                else:
                    obj.set(prop_name, value)
                    logger.info(f"Set Property [{prop_name}]: {value}")

            # Missing Mandatory Reference - Automatic Lookup
            elif prop_desc.is_reference() and prop_desc.is_required():
                try:
                    desc_text = prop_desc.type_description()
                    # Extract type: "List of DataObject of type(s): Fuel" -> "Fuel" or "System" -> "System"
                    if "List" in desc_text:
                        ref_target_type = desc_text.split()[-1][:-1] # "List of DataObject of type(s): Fuel" -> Fuel
                    else:
                        ref_target_type = desc_text.strip() # System 
                    found_refs = STUDY.find(ref_target_type)
                    if found_refs:
                        ref = found_refs[0] if isinstance(found_refs, list) else found_refs
                        logger.info(f"Tryin to auto-resolve mandatory ref adding {ref}") 
                        _set_reference_property(obj, prop_name, prop_desc, ref)
                        logger.warning(f"Auto-assigned default reference for required field '{prop_name}'")
                except Exception as e:
                    logger.error(f"Could not auto-resolve mandatory reference '{prop_name}': {e}")


        logger.info(
            f"SUCCESS: Created '{obj_type}' | "
            f"Identity: [Name: {obj.name}, Code: {obj.code}, Key: {obj.key}] | "
            f"Refs: {obj.get("RefFuels")} | {obj.get("RefSystem")} "
        )
       
        STUDY.add(obj)
        return obj

    except Exception as e:
        logger.critical(f"Failed to create element of type {obj_type}: {str(e)}")
        raise RuntimeError(f"Object creation failed: {e}")
```

```python
def delete_element(obj_key):
    obj = STUDY.get_by_key(obj_key)
    refs = obj.referenced_by()

    if not refs: #Remove directly
        STUDY.remove(obj)
    else: 
        print(f"To remove element {obj} the following elements will be removed to {refs}. Do you want to proceed?")
        for obj_ref in refs: 
            STUDY.remove(obj_ref)
        STUDY.remove(obj)
    

print(delete_element("System System 1 [1]"))
```
