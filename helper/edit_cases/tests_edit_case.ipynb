```python
def get_available_objects(obj_type:str):
    description = f"Available Names/ids for {obj_type}: "
    name_id_map = {}
    for obj in study.find(obj_type):
        name = obj.name.strip()
        id = None
        name_id_map[name] = id
    
    return description + str(name_id_map)

print(get_available_objects("ThermalPlant"))
```

```python
def is_dataframe(obj,key): 
    """Check if the property is a dataframe"""
    description = obj.description(key)
    if description.is_dynamic():
        return True
    if len(description.dimensions()) > 0: 
        return True
    return False
```

```python
def modify_element(obj_id,property:str, value):
    obj = study.find("ThermalPlant")[0]
    obj.set(property,value)
    return obj.get(property)

print(modify_element("","GerMin",6))
```

```python
# Será mais facilmente implementada com o unique id 
def create_modification(obj_id,property:str, modifications: dict):
    obj = study.find("ThermalPlant")[0]
    description = obj.description(property)
    if description.is_dynamic(): 
        for date,value  in modifications.items():
            obj.set_at(property,date,value)

        return True

    else: 
        raise ValueError

modifications = {"02/2013":3,"03/2013":4}
create_modification("","GerMin",modifications)
```
